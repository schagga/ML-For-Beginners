```python
import kagglehub
import pandas as pd

path = kagglehub.dataset_download("abhishek14398/salary-dataset-simple-linear-regression")

print("Path to dataset files:", path)
data = pd.read_csv(path + "/Salary_dataset.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.loc[:, ['YearsExperience', 'Salary']]
print(data)
```

```python
import matplotlib.pyplot as plt

plt.scatter('YearsExperience','Salary',data=data)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
X = data['YearsExperience'].to_numpy().reshape(-1,1)
y = data['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
import numpy as np

model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

mse = np.sqrt(mean_squared_error(y_test, predictions))
print(f'Mean error: {mse:3.3} ({mse/np.mean(predictions)*100:3.3}%)')
```
